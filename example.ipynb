```python
# Importing libraries
import random
import numpy as np
import pandas as pd
from string import ascii_lowercase, digits
```

```python
# creating random data for creating a dataframe
query = ['refregirator', 'watch', 'apple', 'phone', 'chocoloate', 'detergent', "keys", "icecream", "desk", "pen"]
chars = ascii_lowercase + digits
ground_truth_items = [[''.join(random.choice(chars) for _ in range(8)) for _ in range(24)] for _ in range(len(query))]
ground_truth_scores = [sorted([np.round(random.random(), 2) for _ in range(24)],reverse = True ) for _ in range(len(query))]
predicted_items = [random.sample(i, 24) for i in ground_truth_items]
```

```python
# Converting the dictionary to dataframe
df = pd.DataFrame({"query" : query, 
                   "ground_truth_items" : ground_truth_items, 
                   "ground_truth_score" : ground_truth_scores,
                   "predicted_items" : predicted_items})
df.head(5)
```

```python
from rank_validation.validation_generator import get_metrics_report
query_report, overall_report = get_metrics_report(df, 
                                                  "ground_truth_items", 
                                                  "ground_truth_score", 
                                                  "predicted_items", 
                                                  ["ndcg", "kendall_tau", "recall"], 
                                                  [4, 8, 24])
```
